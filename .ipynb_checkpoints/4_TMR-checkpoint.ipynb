The method proposed by the article aims to remove artefacts created by color and contrast modifications on images (such as color transfer, histogram equilization...). 4 types of artefacts are highlighted by the article :
- Noise enhancement
- Compression artefacts :
- Detail loss : mainly due to contrast reduction
- Color proportion inconstencies : occurs when similar colors in image $u$ are no longer similar in image $T(u)$. This happens when modifications between $u$ and $T(u)$ are important (especially with color transfer)

The transportation map is denoted by $\mathcal{M}(u)=T(u)-u$. The article applies the operator $Y_u$ (defined below) to the transportation map (called **Transportation Map Reguralization** (**TMR**)).  
Formally speaking, an image is defined as $v : \Omega=[1,N]\times [1,M]\longrightarrow\mathbb{R}^d$ (where $d$ is the number of channels).

$$Y_u(v):x\in\Omega\mapsto\frac{1}{C(x)}\int_{y\in\mathcal{N}(x)}v(y).w_u(x,y)dy$$

with $w_u(x,y)=e^{-\frac{\|u(x)-u(y)\|^2}{\sigma^2}}$, $\mathcal{N}(x)$ a neighborhood of $x$ and $C(x)=\int_{y\in\mathcal{N}(x)}w_u(x,y)dy$ a normalizaion constant.

The hyperparemeters are $\sigma$ and the size of $\mathcal{N}(x)$. As the authors of the article don't mention the chosen size of the neightborhood, it is arbitrarly set to a $15\times 15$ square centered on $x$

## Naive approach

Here are implemented the formulas mentioned above, as well as iterated TMR in a recursive way.  
However, as one iteration of TMR takes a long time to compute, iterated TMR computation is intractable here for high resolution image as it requires too much memory.

```python
def w(u, x, y, sigma=50):
    return np.exp(- (np.linalg.norm(u[x]-u[y])**2 / sigma**2))


def Y(u, v, N=15):
    res = np.zeros_like(v)
    r, c,_ = v.shape
    half_N = N//2
    for i in tqdm(range(half_N, r-half_N+1)):
        for j in range(half_N, c-half_N+1):
            x = (i, j)
            nbh = v[i-half_N:i+half_N+1, j-half_N:j+half_N+1] # neighborhood
            C = 0

            for k in range(nbh.shape[0]):
                for l in range(nbh.shape[1]):
                    y = (x[0]+k-half_N, x[1]+l-half_N)
                    w_u = w(u, x, y)
                    res[x] += v[y]*w_u
                    C += w_u
            res[x] /= C
    return res

def TMR(T_u, u):
    return Y(u, T_u) + u - Y(u,u)

def Y_k(u, v, k):
    if k == 0:
      return v
    else:
      v = Y(u,v)
      return Y_k(u,v,k-1)

def iterated_TMR(T_u, u, k):
    return Y_k(u, T_u - u, k) + u
```

```python
k = 3
res = iterated_TMR(im_target, im_src, k)
```

```python
res = (res - res.min()) / (res.max() - res.min())
fig, ax = plt.subplots(1, 3, figsize=(17,5))
ax[0].imshow(im_target)
ax[0].set_title('(a) Raw colored image')
ax[0].axis('off')
ax[1].imshow(res)
ax[1].set_title(rf'(b) Iterated TMR (k={k})')
ax[1].axis('off')
ax[2].imshow(im_src)
ax[2].set_title('(c) Source image')
ax[2].axis('off')
plt.tight_layout()
plt.show()
```

One can notice that image (b) has recovered some details lost during color transfer, especially in the background.  
Moreover, colors saturation on image (a) have completely dissapeared on the result image. 
One notice that image (b) is also closer to the source image (c) in terme of color map. It looks like an average image between (a) and (c).

## Smarter approach (but intractable here)

The second approach consists in applying TMR as a matrix $A$ to an image $u$ reshaped as a vector $v$.  
As the obtained matrix $A$ is stochastic (cf paper), we know that $(A^k)_{k\in\mathbb{N}}$ will converge 

```python
u = im_src
nbh_size=15
r, c = u.shape[:-1]
m = r * c
# v = u.reshape((m,-1))
A = np.zeros((m,m), dtype=np.float16)
half_nbh = nbh_size//2

def compute_coef(args):
    i, j = args
    # get corresponding coordinates in 2D image
    i_2d = np.unravel_index(i, (r, c))
    j_2d = np.unravel_index(j, (r, c))

    # compute min and max indices of a neighborhood centered on i_2d
    i_min, i_max = max(0,i_2d[0]-half_nbh), min(r-1,i_2d[0]+half_nbh+1)
    j_min, j_max = max(0,i_2d[1]-half_nbh), min(c-1,i_2d[1]+half_nbh+1)
    # N_i = u[i_min:i_max, j_min:j_max]
    N_i_idx = [(k, l) for k in range(i_min,i_max) for l in range(j_min,j_max)]

    # check if j 2d index is in the neighborhood
    if j_2d in N_i_idx:
        a=w(u,i_2d,j_2d)
        # compute normalization constant
        weight = 0
        for idx in N_i_idx:
            weight+=w(u,i_2d,idx)

        A[i,j] = a/weight
    else: A[i,j] = 0

with multiprocessing.Pool() as pool:

    args_list = [(i, j) for i in range(m) for j in range(m)]
    pool.starmap(compute_coef, args_list)
```
